```python
from EmotionModels import EmotionClassifier
from bertviz import model_view
```

```python
model_name = "FacebookAI/xlm-roberta-base"
model_path = R"D:\Misc\xlm-roberta1.pth"
```

```python
text = "the quick brown fox jumps over the lazy dog"
```

```python
model = EmotionClassifier.from_trained(model_path, output_attentions=True)
out = model.model.encoder.tokenizer(text, return_tensors="pt", truncation=True)
attentions = model.model.encoder.encoder(out.input_ids, out.attention_mask).attentions
tokens = model.model.encoder.tokenizer.convert_ids_to_tokens(out.input_ids[0])
```

```python
model_view(attentions, tokens)
```

```python
from EmotionModels import EmotionClassifier, EmotionDataset
from bertviz import model_view
from lime.lime_text import LimeTextExplainer
```

```python
DEVICE = "cpu"
model_name = "FacebookAI/xlm-roberta-base"
model_path = R"D:\Misc\xlm-roberta1.pth"

model = EmotionClassifier.from_trained(model_path, output_attentions=True)
model = model.to(DEVICE)
explainer = LimeTextExplainer(class_names=EmotionDataset.labels)
```

```python
text = "help, he is coming after me"
```

```python
exp = explainer.explain_instance(text, model.predict_proba, num_samples=2000, labels=range(len(EmotionDataset.labels)))
exp.show_in_notebook(text=False, labels=None)
```
